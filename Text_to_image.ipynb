```python

from diffusers import DiffusionPipeline
import gradio as gr
import torch

# Initialize the DiffusionPipeline
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

def generate_image(prompt):
    # Generate an image based on the prompt
    images = pipe(prompt=prompt).images[0]

    # Assuming `images` is a PIL image or compatible, you can return it directly.
    return images

# Create Gradio Interface
iface = gr.Interface(
    fn=generate_image,
    inputs=gr.inputs.Textbox(lines=2, placeholder="Type something here..."),
    outputs="image",
    #live=True,
    capture_session=True,
    title="Flamehead Labs - Text to Image Generator",
    description="This interface converts your textual prompt into a corresponding image using advanced AI models. Simply enter your prompt and watch the AI generate an image for you.",
    examples=[
        ["A sunset over the mountains."],
        ["A cat playing with a ball of yarn."]
    ],
    theme="huggingface"  # Assuming "huggingface" is a valid theme, otherwise use any valid theme
)

# Launch the interface
iface.launch()
```
